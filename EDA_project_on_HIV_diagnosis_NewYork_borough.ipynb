# Introduction

In the ever-evolving landscape of public health, understanding the nuanced patterns of disease prevalence and transmission is crucial to capture the effective strategies for prevention and intervention. 
New York city, a metropolis that thrives with diversity and complexity, is no exception to the challenges posed by infectious diseases. 
The objective of this project is to derive into the intricate tapestry of HIV/AIDS diagnoses in New York City, examining how the devastating disease affects different communities across the city. 

We will explore the multifaceted aspects of HIV/AIDS, including its distribution by neighborhood, its impact on various racial and ethnic groups, the role of age, and the variations seen across the city's boroughs. This comprehensive data analysis will not only reveal existing disparities but will also provide insights critical to the design and implementation of targeted, data-driven interventions. By investigating the intersections of neighborhood, race, age, and borough in relation to HIV/AIDS diagnoses, we aim to contribute to the ongoing efforts to combat this epidemic and foster a healthier, more equitable New York City.


These data were reported to the NYC DOHMH by March 31, 2021.
This dataset includes data on new diagnoses of HIV and AIDS in NYC for the calendar years 2010 through 2020. Reported cases and case rates (per 100,000 population) are stratified by United Hospital Fund (UHF) neighborhood, age group, and race/ethnicity.

This dataset last updated march 2023.
[Dataset](https://data.cityofnewyork.us/d/dxnu-p2qd?category=Health&view_name=HIV-AIDS-Diagnoses-by-Neighborhood-Age-Group-and-R)

### Data cleaning

Although we have converted some of our variables to numerical, rename columns, capital to lower case but there are still some irrelevant data that we do not need. We can remove this data using regular expressions.

```python
# view the age column where is some string data exists
hiv_data.age.value_counts()
```

```python
# view the race column where is some '/' exists
hiv_data.race.value_counts()
```

```python
# view the neigborhood column where is some '-' data exists
hiv_data.neighborhood.value_counts()[-10:]
```

__*Using Regular Expression method we will clean all unwanted string, /, - , \n from the data to make more accurate our further analysis*__ 

```python
# Use str.extract to extract the age range
hiv_data['age'] = hiv_data['age'].replace(r'[^0-9-]', '', regex=True)
hiv_data['race'] = hiv_data['race'].str.split('/').str[0]
hiv_data['neighborhood'] = hiv_data['neighborhood'].str.split('-').str[0]
hiv_data['borough'] = hiv_data['borough'].str.replace('\n', ' ')
hiv_data['total_no_of_hiv_diagnoses'] = hiv_data['total_no_of_hiv_diagnoses'].replace(r'[^0-9-]', '', regex=True)
hiv_data['hiv_diagnoses_per_100,000_population'] = hiv_data['hiv_diagnoses_per_100,000_population'].replace(r'[^0-9-]', '', regex=True)
hiv_data['total_no_of_concurrent_hiv/aids_diagnoses'] = hiv_data['total_no_of_concurrent_hiv/aids_diagnoses'].replace(r'[^0-9-]', '', regex=True)
hiv_data['proportion_of_concurrent_hiv/aids_diagnoses_among_all_hiv_diagnoses'] = hiv_data['proportion_of_concurrent_hiv/aids_diagnoses_among_all_hiv_diagnoses'].replace(r'[^0-9-]', '', regex=True)
hiv_data['total_no_of_aids_diagnoses'] = hiv_data['total_no_of_aids_diagnoses'].replace(r'[^0-9-]', '', regex=True)
hiv_data['aids_diagnoses_per_100,000_population'] = hiv_data['aids_diagnoses_per_100,000_population'].replace(r'[^0-9-]', '', regex=True)

# Check the first 5 rows after cleaning some string data
hiv_data.head()
```

```python
# Show the age column now after cleaning
hiv_data.age.value_counts()
```

The age column has now been cleaned to remove newlines and duplicate indices. To make the age column more informative, we define age groups with descriptive names, such as "kids" (0-13), "teenage" (13-19), "young" (20-29), and "adult" (30-39). I have created a lambda function to perform this task, which is shown in the cell below.

```python
# Create a lambda function to categorize the age ranges
    
categorize_age = lambda x: 'kids' if '0-13' in x else'teenage' if '13-19' in x else'young' if '20-29' in x else 'adult' if '30-39' in x else 'middle age' if '40-49' in x else 'upper age' if '50-59' in x else 'old'

# Apply the lambda function to create a new 'age_category' column
hiv_data['age_category'] = hiv_data['age'].apply(categorize_age)
hiv_data.age_category.value_counts()
```

```python
# Check where we have removed all unwanted text
print(hiv_data.race.value_counts())
print("---------------------------")

print(hiv_data.borough.value_counts())
print("---------------------------")

print(hiv_data.neighborhood.value_counts()[-10:])
```

After removing all unnecessary characters, the data is now much cleaner. However, the value "ALL" is still present in the data. This value is redundant, as we already have separate values for borough, race, neighborhood, and age category. Additionally, there are no data points associated with the value "ALL". Therefore,the value "ALL" will be removed from the data.

```python
# Remove rows with a specific value 'All'in few columns
value_to_remove = 'All'
hiv_data = hiv_data.drop(hiv_data[hiv_data['borough'] == value_to_remove].index)

hiv_data = hiv_data.drop(hiv_data[hiv_data['race'] == value_to_remove].index)
hiv_data = hiv_data.drop(hiv_data[hiv_data['neighborhood'] == value_to_remove].index)
```

```python
# Check where we have removed the all
print(hiv_data.race.value_counts())
print("---------------------------")

print(hiv_data.borough.value_counts())
print("---------------------------")

print(hiv_data.neighborhood.value_counts()[:5])
```

__*Data is almost clean now, we have checked previously that there is no 'All', unwanted string value. Data type has been modified, age column has been classified age group. Now have to find the missing value, statistical description and remove the null values or fill the null values by fill method in pandas to do better calculation, better visualization.*__

```python
# Find the statistical description of all variable in in hiv dataset
hiv_data.describe(include = "all")
```

Descriptive statistics, such as the minimum, maximum, and mean values, can be calculated for numerical variables, and the top value, unique number, and frequency count can be calculated for categorical variables. However, the presence of missing values in the dataset reduces the accuracy of these results. 

In this step, we will impute the missing values in the dataset using the mean for numerical variables and the mode for categorical variables. Imputation is the process of replacing missing values with estimated values. The mean and mode are the most efficient imputation methods for numerical and categorical variables, respectively, because they minimize the bias in the data. 

```python
# for numerical data fill with mean
hiv_data1 = hiv_data.fillna(hiv_data.mean())
hiv_data1.isna().sum()
```

```python
# for categorical data fill with mode
hiv_data1.fillna(hiv_data1['borough'].mode()[0], inplace=True)
hiv_data1.isna().sum()
```

Finally, we will remove the outliers from numerical data variables by calculating the mean and standard deviation.

```python
# Filter only the numerical columns
numerical_columns = hiv_data1.select_dtypes(include=[float])

# Calculate the mean and standard deviation for each column
means = numerical_columns.mean()
stds = numerical_columns.std()

# Define a threshold (e.g., 2 standard deviations) for outlier removal
threshold = 3

# Identify and remove outliers for each column
df_no_outliers = numerical_columns[(np.abs(numerical_columns - means) <= threshold * stds).all(axis=1)]

# Print the DataFrame without outliers
df_no_outliers.head()
```

We calculate the means and standard deviations for each column.
We define a threshold (e.g., 3 standard deviations) to identify outliers.
We create a new DataFrame, df_no_outliers, by selecting rows where the absolute differences between the values and their respective column means are within the threshold for all columns (using .all(axis=1)).
df_no_outliers will contain the data with outliers removed for all numerical columns.
